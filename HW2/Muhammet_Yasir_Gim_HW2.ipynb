```python
import os

from bs4 import BeautifulSoup

from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import pandas as pd
```

```python
user_name = os.environ.get('USER')
WORKDIR = f"/Users/{user_name}"
DRIVER_PATH = WORKDIR +  "/Downloads/chromedriver"

options = Options()
options.add_argument("--window-size=1920,1200")

driver = webdriver.Chrome(options=options, executable_path=DRIVER_PATH)

categories = [
    {"name": "Atıştırmalık", "url": "https://www.a101.com.tr/market/atistirmalik/?page=", "max_page":27},
    {"name": "Ev Bakım & Temizlik", "url": "https://www.a101.com.tr/market/ev-bakim-temizlik/?page=", "max_page":11},
    {"name": "İçecek", "url": "https://www.a101.com.tr/market/icecek/?page=", "max_page":15},
    {"name": "Kahvaltılık & Süt Ürünleri", "url": "https://www.a101.com.tr/market/kahvaltilik-sut-urunleri/?page=", "max_page":10},
    {"name": "Ambalaj Malzemeleri", "url": "https://www.a101.com.tr/market/ambalaj-malzemeleri/?page=", "max_page":2},
    {"name": "Hazır Yemek", "url": "https://www.a101.com.tr/market/hazir-yemek/?page=", "max_page":4},
    {"name": "Meyve & Sebze", "url": "https://www.a101.com.tr/market/meyve-sebze/?page=", "max_page":2},
    {"name": "Et & Tavuk & Balık", "url": "https://www.a101.com.tr/market/et-tavuk-balik/?page=", "max_page":2},
    {"name": "Sağlıklı Yaşam Ürünleri", "url": "https://www.a101.com.tr/market/saglikli-yasam-urunleri/?page=", "max_page":3},
    {"name": "Temel Gıda", "url": "https://www.a101.com.tr/market/temel-gida/?page=", "max_page":16}
]

item_prices = []

for category in categories:
    max_page_number = category["max_page"]
    category_name = category["name"]
    for i in range(1, max_page_number):

        driver.get(category["url"] + str(i))
        WebDriverWait(driver, 500)

        responses.append(driver.page_source)

        for response in responses:
            try:
                soup = BeautifulSoup(response)
                name = "li"
                class_name = "col-md-4 col-sm-6 col-xs-6 set-product-item"

                item_list = soup.find_all(name, {"class": class_name})
                print(len(item_list))
                for item in item_list:
                    price = float(item.find("span", {"class": "current"}).text[1:].strip().replace(",", "."))
                    item_name = item.find("h3", {"class": "name"}).text.strip()
                    item_prices.append([item_name,category_name, price])

                responses.clear()

            except Exception as e:
                print(e)
        
driver.quit()
```
